```python
# 18번
class Animal:
  def __init__(self, name):
    self.name = name
  def sound(self):
    return 'sound'

class Cat(Animal):
  def __init__(self, name, level):
    super().__init__(name)
    self.level = level
  def sound(self):
    return 'walwal'

o = Cat('Wash','Top')
print(f'{o.name} : {o.sound()} ({o.level})')
```

```python
# 19번
def ch(val):
  if type(val) == type(""):
    return len(val)
  elif type(val) == type(1):
    return 1
  elif type(val) == tuple:
    return 2
  elif type(val)  == list:
    return 3
  else:
    return 4

a = "5"
b = 200
c = (1,)
d = [1,2]
print( ch(a), ch(b), ch(c), ch(d))
```

```python
# 20번
def sear(t, c):
  p = []
  for i in range(len(t)):
    if t[i] == c:
      p.append(i)
  return p
t = 'Life is too short'
c = 's'
print(f'{c} : {sear(t,c)}')
```

```python
# 21번
def ch_cnt(st):
  cnt = {}
  for ch in st:
    if ch in cnt:
      cnt[ch] += 1
    else:
      cnt[ch] = 1

  return cnt
str = 'Pass'
print(f'cnt : {ch_cnt(str)}')
```

```python
# 22번
def change(v, t=(1,)):
  t = t + (v,)
  return t

print(change(2))
print(change(3, (5,4,3)))
```

```python
# 23번
n = 10
def inc():
  n = 5
  print(f'inc : {n}')
def dec():
  global n
  n -= 5
  print(f'dec : {n}')

print(f'inc() : {inc()}')
print(f'n : {n}')
dec()
print(f'n : {n}')
```
