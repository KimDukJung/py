**클래스**

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def disp(self):
    print(f'이름: {self.name}, 나이: {self.age}')

  def get_name(self):
    return self.name

  def get_age(self):
    return self.age

p1 = Person('Kim', 30)
p2 = Person('Lee',25)
p1.disp()
print(p2.get_name(), p2.get_age())
```

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def addAge(self):
    self.age += 1

  def disp(self):
    return f'{self.name}, {self.age}'

p = Person('kim', 30)
p.addAge()
print(p.disp())
p.addAge()
print(p.disp())
```

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def res(self):
    if self.age >= 30:
      return f'{self.name} Young'
    else:
      return f'{self.name} Very Young'

p1 = Person('kim', 40)
p2 = Person('Lee', 25)
print(p1.res())
print(p2.res())
```

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def prn(self):
    return f'{self.name}({self.age})'

p1 = Person('kim', 40)
p2 = Person('Lee', 25)
p3 = Person('Park', 20)
persons = [p1, p2, p3]
for p in persons:
  print(p.prn())
```

```python
class Counter:
  count = 0

  def __init__(self, name):
    self.name = name
    Counter.count += 1

c1 = Counter('kim')
c2 = Counter('lee')
c3 = Counter('park')
print(Counter.count)
```

**클로저**

```python
def outer():
  n = 10
  def inner():
    return n + 5
  return inner
f = outer()
print(f())
```

```python
def outer():
  msg = 'Hi'
  def inner():
    return msg + ' Hello'
  return inner

fn = outer()
print(fn())
```

```python
def outer():
  a=5
  def inner():
    return a + 2
  return inner

fa = outer()
fb = outer()
print(fa(), fb())
```

```python
def outer(n):
  def inner():
    return n + 2
  return inner

fa = outer(5)
fb = outer(10)
print(fa(), fb())
```
